```python
def issafe(row,col,board,n):
    i=row-1
    while i>=0:
        if board[i][col]==1:
            return False
        i-=1
        
    i=row-1
    j=col-1
    while i>=0 and j>=0:
        if board[i][j]==1:
            return False
        i-=1
        j-=1
    i=row-1
    j=col+1
    while i>=0 and j<n:
        if board[i][j]==1:
            return False
        i-=1
        j+=1
    return True

def printpathhelper(row,n,board):
    if row==n:
        for i in range(n):
            for j in range(n):
                print(board[i][j],end=' ')
        print()
        return
    
    for col in range(n):
        if issafe(row,col,board,n) is True:
            board[row][col]=1
            printpathhelper(row+1,n,board)
            board[row][col]=0
    return

def printpath(n):
    board=[[0 for j in range(n)]for i in range(n)]
    printpathhelper(0,n,board)

n=int(input())
printpath(n)
```

```python
def pir(b,r,n):
    for j in range(9):
        if b[r][j]==n:
            return True 
    return False

def pic(b,c,n):
    for j in range(9):
        
        if b[j][c]==n:
            return True 
    return False

def pib(board,row,col,n):
    for i in range(row,row+3):
        for j in range (col,col+3):
            if board[i][j]==n:
                return True 
    return False

def  ispossible(b,r,c,n):
    if(pir(b,r,n)):
        return False
    if(pic(b,c,n)):
        return False
    if(pib(b,r-(r%3),c-(c%3),n)):
        return False
    return True



    
def solveSudoku(b):
    r=-1
    c=-1
    flag = False
    for i in range(9):
        for j in range(9):
            if b[i][j] is 0:
                r=i
                c=j
                flag=True
                break
        if flag is True:
            break
    if r == -1:
        return True
    for n in range(1,10):
        if(ispossible(b,r,c,n)):
            b[r][c]=n
            if(solveSudoku(b)is True):
                return True
            b[r][c]=0
    return False
   

board = [[ int(ele) for ele in input().split() ]for i in range(9)]
ans = solveSudoku(board)
if ans is True:
    print('true')
else:
    print('false')
```
