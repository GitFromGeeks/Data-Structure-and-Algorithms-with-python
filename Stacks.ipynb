```python

from sys import stdin

#Following is the structure of the node class for a Singly Linked List
class Node :

    def __init__(self, data) :
        self.data = data
        self.next = None


class Stack :

    #Define data members and __init__()
    def __init__(self):
        self.__head = None
        self.__c = 0

    '''----------------- Public Functions of Stack -----------------'''


    def getSize(self) :
        return self.__c



    def isEmpty(self) :
        return self.getSize() ==0



    def push(self, data) :
        newNode = Node(data)
        newNode.next = self.__head
        self.__head = newNode
        self.__c += 1



    def pop(self) :
        if self.isEmpty() is True:
            return -1
        
        data = self.__head.data
        self.__head = self.__head.next
        self.__c -= 1
        
        return data



    def top(self) :
        if self.isEmpty():
            return -1
        return self.__head.data



#main
q = int(stdin.readline().strip())

stack = Stack()

while q > 0 :

    inputs = stdin.readline().strip().split(" ")
    choice = int(inputs[0])

    if choice == 1 :
        data = int(inputs[1])
        stack.push(data)

    elif choice == 2 :
        print(stack.pop())

    elif choice == 3 :
        print(stack.top())

    elif choice == 4 : 
        print(stack.getSize())

    else :
        if stack.isEmpty() :
            print("true")
        else :
            print("false")

    q -= 1
```

```python

def checkBalanced(s):
    dic={"}":"{",")":"(","]":"["}
    tmp=[]
    for x in s:
        if x=="{" or x=="(" or x=="[" or x=="}" or x==")" or x=="]" :
            if tmp and tmp[-1]==dic.get(x):
                tmp.pop()
            else:
                tmp.append(x)
    return tmp
exp=input()
if checkBalanced(exp):
    print('false')
else:
    print('true')

```

```python

def reverseStack(s1,s2):
    if len(s1) <= 1:
        return 
    
    
    while len(s1) != 1:
        ele = s1.pop()
        s2.append(ele)
        
    lastElement = s1.pop()
    while len(s2) != 0:
        ele = s2.pop()
        s1.append(ele)
    reverseStack(s1, s2)
    
    s1.append(lastElement)

from sys import setrecursionlimit
setrecursionlimit(11000)
n = int(input())
if n!=0:
    s1 = [int(ele) for ele in input().split()]
    s2 = []
    reverseStack(s1,s2)
    while(len(s1) !=0):
        print(s1.pop(),end= ' ')
    
```

```python

from sys import stdin

def checkRedundantBrackets(expression) :
    stk = list()
    for i in range(len(expression)) :
        if (expression[i] == '(') or (find(expression[i])) :
            stk.append(expression[i])
        elif expression[i] == ')' :
            hasOperator = False
            while not isEmpty(stk) and top(stk) != '(' :
                stk.pop()
                hasOperator = True
            if not hasOperator :
                return True
            if not isEmpty(stk) :
                stk.pop()
    return False

def find(ch) :
    if ch == '+' or ch == '-' or ch == '*' or ch == '/' :
        return True
    return False

def isEmpty(stack) :
    return len(stack) == 0

def top(stack) :
    return stack[len(stack) - 1]


expression = stdin.readline().strip()

if checkRedundantBrackets(expression) :
	print("true")
else :
	print("false")
```

```python
def countBracketReversals(string):
    if(len(string) == 0):
        return 0
    if(len(string)%2 != 0):
        return -1
    s = []
    for char in string:
        if char == '{':
            s.append(char)
        else:
            if(len(s) > 0 and s[-1] == '{'):
                s.pop()
            else:
                s.append(char)
    count = 0
    while(len(s) != 0):
        c1 = s.pop()
        c2 = s.pop()
        if c1!=c2:
            count+=2
        else:
            count+=1
    return count
string = input()
ans = countBracketReversals(string)
print(ans)



```
